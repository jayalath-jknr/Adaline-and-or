If we consider I1 I2 columns as feature sets, we can add I1, I2 columns as feature sets, but as

```python
or_features=np.array([[-1,-1],[-1,1],[1,-1],[1,1]]) #creating the dataset for training the model
or_target=np.array([-1,1,1,1]) #target variable values
```

```python
#initializing the hyperparameters
weight=[0.5,0.5]
bias=0.1
learning_rate=0.2
epoch=10
```

```python
sum_squared_error=0.0 #to see the overall error of the learned model

for i in range(epoch): #iterating through the number of times the algorithm should run through the dataset

  for j in range(or_features.shape[0]): #iterate through all possible combinations of inputs

    actual_output=or_target[j] # obtain the corresponding target value

    #obtain the values in input combination
    X1=or_features[j][0]
    X2=or_features[j][1]

    #calculate the net unit value to calculate the error
    unit=(X1 * weight[0]) + (X2 * weight[1]) + bias

    #error is calculated for the current iteration
    error=actual_output-unit

    print("error =", error)

    #update the current sum of squared error value using the current error value of the current iteration
    sum_squared_error += error*error

    #update the weights to further reduce the sum squared error based on the learning rate and current iteration error
    weight[0] += learning_rate*error*X1
    weight[1] += learning_rate*error*X2

    #update bias based on current iteration
    bias += learning_rate*error

  print("sum of squared error = ", sum_squared_error/4, "\n\n")
```

AND GATE

Same method which was used to teach the OR gate can be used to train the algorithm so it can learn the AND gate

```python
and_features=np.array([[-1,-1],[-1,1],[1,-1],[1,1]]) #creating the dataset for training the model
and_target=np.array([-1,-1,-1,1]) #target variable values
```

```python
and_weight=[0.5,0.5]
and_bias=0.1
and_learning_rate=0.2
and_epoch=10
```

```python
and_sum_squared_error=0.0 #to see the overall error of the learned model

for i in range(and_epoch): #iterating through the number of times the algorithm should run through the dataset

  for j in range(and_features.shape[0]): #iterate through all possible combinations of inputs

    actual_output=and_target[j] # obtain the corresponding target value

    #obtain the values in input combination
    X1=and_features[j][0]
    X2=and_features[j][1]

    #calculate the net unit value to calculate the error
    unit=(X1 * and_weight[0]) + (X2 * and_weight[1]) + and_bias

    #error is calculated for the current iteration
    error=actual_output-unit

    print("error =", error)

    #update the current sum of squared error value using the current error value of the current iteration
    and_sum_squared_error += error*error

    #update the weights to further reduce the sum squared error based on the learning rate and current iteration error
    and_weight[0] += and_learning_rate*error*X1
    and_weight[1] += and_learning_rate*error*X2

    #update bias based on current iteration
    and_bias += and_learning_rate*error

  print("sum of squared error = ", and_sum_squared_error/4, "\n\n")
```

The above method can be used to train an Adaline to learn any logic gate

```python
# import the module numpy
import numpy as np


# the features for the or model , here we have
# taken the possible values for combination of
# two inputs
features = np.array(
	[
		[-1, -1],
		[-1, 1],
		[1, -1],
		[1, 1]
	])


# labels for the or model, here the output for
# the features is taken as an array
labels = np.array([-1, 1, 1, 1])

# to print the features and the labels for
# which the model has to be trained
print(features, labels)

# initialise weights, bias , learning rate, epoch
weight = [0.5, 0.5]
bias = 0.1
learning_rate = 0.2
epoch = 10

for i in range(epoch):

	# epoch is the number of the model is trained
	# with the same data
	print("epoch :", i+1)

	# variable to check if there is no change in previous
	# weight and present calculated weight
	# initial error is kept as 0
	sum_squared_error = 0.0

	# for each of the possible input given in the features
	for j in range(features.shape[0]):

		# actual output to be obtained
		actual = labels[j]

		# the value of two features as given in the features
		# array
		x1 = features[j][0]
		x2 = features[j][1]

		# net unit value computation performed to obtain the
		# sum of features multiplied with their weights
		unit = (x1 * weight[0]) + (x2 * weight[1]) + bias

		# error is computed so as to update the weights
		error = actual - unit

		# print statement to print the actual value , predicted
		# value and the error
		print("error =", error)

		# summation of squared error is calculated
		sum_squared_error += error * error

		# updation of weights, summing up of product of learning rate ,
		# sum of squared error and feature value
		weight[0] += learning_rate * error * x1
		weight[1] += learning_rate * error * x2

		# updation of bias, summing up of product of learning rate and
		# sum of squared error
		bias += learning_rate * error

	print("sum of squared error = ", sum_squared_error/4, "\n\n")
```
